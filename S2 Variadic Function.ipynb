Take multiple postional arguments using *args
*args - will act like tuple

```python
# Variadic function with positional arguments
def func1(*args):
    print(f"Postional Arguments : {args}")
    print(f"Type : {type(args)}")
    for i in args:
        print(i)
```

```python
func1(2, 3)
```

```python
func1("Test", 101, 102, 34.5, 46, "Example")
```

```python
func1(3, 4, 5)
```



Write a variadic function to calculate sum of any given numbers

```python
def add_many(*args):
    # Initialize sum to 0
    s = 0
    # Apply for loop on args
    for i in args:
        s = s + i
    # Return the value of sum
    return s
```

```python
add_many(3, 4)
```

```python
add_many(1, 2, 3)
```

```python
add_many(5, 6, 7, 8, 9)
```

```python
# You can pass list as positional argument in above
a = [23, 24, 25, 11, 12, 13, 14, 15]
add_many(*a)
```

```python
b = list(range(1, 101))
print(b)
```

```python
add_many(*b)
```

Create a varidic function to calculate product of numbers (multiplication)

```python
def multiply_many(*args):
    # Intitialize product with 1
    p = 1
    # Apply for loop on args
    for i in args:
        p = p * i
    # Return value of product
    return p
```

```python
multiply_many(3, 4)
```

```python
multiply_many(12, 13, 14)
```

```python
multiply_many(45, 11, 12, 13, 18, 19)
```

```python
# Whenever providing a list unpack with *list
c = [3, 4, 5, 11, 12, 13]
multiply_many(*c)
```

Write a varidic function to calcualte the cart final bill

```python
cart1 = [("Mouse", 2, 600), ("Keyboard", 1, 1200), ("Laptop", 2, 35000)]
```

```python
def billing_var(*cart):
    # Intitilize total bill to 0
    total_bill = 0
    # Apply for loop on the cart value
    for name, qty, price in cart:
        line_total = qty * price
        total_bill = total_bill + line_total
        print(
            f"Product : {name}, Quantity : {qty}, Price : {price} INR, Line Total : {line_total}"
        )
        print("=" * 60)
    # Return the total bill value
    return total_bill
```

```python
cart1
```

```python
bill1 = billing_var(*cart1)
print(f"Total Bill : {bill1} INR")
```

```python
cart2 = [
    ("Mobile Cover", 5, 100),
    ("Screen Guard", 10, 200),
    ("Wireless Mouse", 1, 1500),
    ("Keyboard", 1, 2000),
    ("HDMI Cable", 2, 300),
]
bill2 = billing_var(*cart2)
print(f"Total Bill : {bill2} INR")
```

```python
bill3 = billing_var(("Screen", 15000, 2), ("Microphone", 3000, 3))
print(f"Total bill : {bill3} INR")
```

```python
bill4 = billing_var(
    ("USB Drive", 4, 500),
    ("Earphones", 1, 1200),
    ("Keyboard", 1, 1500),
    ("Screen Guard", 2, 300),
)
print(f"Total bill : {bill4} INR")
```

**kwargs - act like dictionary

```python
def func2(**kwargs):
    print(f"Keyword arguments : {kwargs}")
    print(f"Type : {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"Key : {key}, Value : {value}")
```

```python
func2(name="Rahul", age=23)
```

```python
func2(name="Aditi", age=25, email="a@test.com", weight=45.3)
```

```python
func2(food_name="Paneer", price="200", quantity=5, extra="spicy")
```

```python
# You can unpack the dictionary **dict
a = {"emp_id": 101, "name": "Rahul", "age": 28, "dept": "Engg."}
func2(**a)
```

Invoice generation **kwargs

```python
def generate_invoice(customer, **fields):
    invoice = {"customer": customer}
    invoice.update(fields)
    return invoice
```

```python
inv1 = generate_invoice(
    "Sarthak",
    products=["Laptop", "Mouse", "Keyboard"],
    amount=75000,
    cgst=5000,
    sgst=5000,
    total_bill=80000,
    paid=True,
)
inv1
```

Combine the *args and **kwargs
*args - Tuple, **kwargs - dictionary



```python
def func3(*args, **kwargs):
    print(f"Postional Arguments : {args}")
    print(f"Type : {type(args)}")
    for i in args:
        print(i)
    print(f"Keyword arguments : {kwargs}")
    print(f"Type : {type(kwargs)}")
    for key, value in kwargs.items():
        print(f"Key : {key}, Value : {value}")
```

```python
func3("Example", 23, 11, name="Rahul", age=23, dept="Engg.", height=5.5)
```

```python
# Unpack the list *list, dictionary with **dict
a = [12, 13, 14, 15, 16]
b = {"food_name": "Panner", "price": 250}
func3(*a, **b)
```

```python
func3(12, 13, 14)
```

```python
func3(a=12, b=13, c=25, d=11)
```

Create a user having differnt roles and details

```python
def create_user(username, *roles, **details):
    user = {"username": username}
    user["roles"] = list(roles)
    user.update(details)
    return user
```

```python
u1 = create_user("sarthaks1", "admin", "editor", city="Pune", dept="Engg.", age=28)

u1
```
