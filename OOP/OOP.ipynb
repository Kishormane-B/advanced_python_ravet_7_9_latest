```python
class Person:

    def __init__(self, name: str, age: int, gender: str, occupation: str):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

    def intro(self):
        print(f"My name is {self.name}")
        print(f"My age is {self.age} years old")
        print(f"I am {self.gender}")
        print(f"I work as {self.occupation}")
```

Create a bank account class


```python
class BankAccount:

    def __init__(self, ac_no: int, name: str, balance: float = 0.0):
        self.ac_no = ac_no
        self.name = name
        self.balance = balance
        if balance < 0:
            raise ValueError("Negative Balance is Invalid")

    def __str__(self):
        return f"BankAccount(ac_no = {self.ac_no}, name = {self.name})"

    def get_balance(self):
        print(f"Account No : {self.ac_no}, Balance : {self.balance} INR")

    def deposit(self, amt: float):
        if amt <= 0:
            raise ValueError("Amount cannot be negative or zero")
        self.balance = self.balance + amt

    def withdraw(self, amt: float):
        if amt <= 0:
            raise ValueError("Amount cannot be negative or zero")
        if self.balance < amt:
            raise ValueError("Insufficient balance")
        self.balance = self.balance - amt

    def transfer(self, amt: float, other_ac: "BankAccount"):
        self.withdraw(amt)
        other_ac.deposit(amt)
```
